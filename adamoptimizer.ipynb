```python
import pandas as pd
import numpy as np
import random
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
df.astype(float)
```

```python
classlabel= df.pop("Outcome") 
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test = train_test_split(df,classlabel,test_size=0.3,random_state=42)
```

```python
model = Sequential([
    Dense(8, input_shape=(8,), activation='sigmoid'),
    Dense(6, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False),
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train.values, y_train.values, epochs=1000)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
```

```python
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print('\nTest accuracy: ', test_acc)
print('\nTest loss: ', test_loss)
```
