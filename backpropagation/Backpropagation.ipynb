```python
df=df.astype('float128')
```

```python
target_output=np.array(df["Outcome"])
target_output
```

```python
df=df.drop(columns=["Outcome"])
```

```python
df.dtypes
```

```python
input_features=[]
for i in range(len(df)):
  input_features.append(df.loc[i])
input_features=np.array(input_features)
```

```python
input_features.shape
input_features.dtype
```

```python
w_ji = np.random.uniform(low=-0.02, high=0, size=(8,6))
w_ji.astype('float128')
```

```python
w_ji.shape
```

```python
w_kj = np.random.uniform(low=-0.02, high=0, size=(6,1))
w_kj.astype('float128')
```

```python
w_kj.shape
```

```python
lr=0.001
def sigmoid (x):
    return 1/(1+np.exp(-x))
```

```python
for epoch in range(150000):
    net_j=np.dot(input_features,w_ji)
    oj= sigmoid(net_j)
    net_k=np.dot(oj,w_kj)
    ok=sigmoid(net_k)

    error_out=((1/2)* (np.power((ok-target_output),2)))
    print(error_out.sum())

    #phase 1
    delt_k=(ok*(1-ok))*(ok - target_output)
    z1=np.dot(oj.T, delt_k)

    #phase 2
    s=np.dot(delt_k, w_kj.T)
    delt_j=(oj*(1-oj))*s
    z2=np.dot(input_features.T,delt_j)

    w_kj=w_kj-lr*z1
    w_ji=w_ji-lr*z2
```

```python
p=np.array([1	,126,	60	,0,	180	,30.1	,0.349,	47	],dtype=np.float128)
r1=np.dot(p,w_ji)
r2=sigmoid(r1)
r3=np.dot(r2,w_kj)
r4=sigmoid(r3)
print(r4)
```
